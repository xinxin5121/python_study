```python
def start():
    player1 = [1,1]
    #player2 is computer
    player2 = [1,1]
    player = [player1, player2]
    return player
```

```python
def decide_turn(count):
    p = 0 if count%2 == 1 else 1
    return p
```

```python
def input_heli(player, p, count):
    #可以直接默认player1的 但我懒得改了
    current_player = player[p]
    if len(current_player) == 2:
        player_input = eval(input("\nRound %s: \nPlayer %s's turn \nChoose from: %s \nEnter number:" % \
                           (count, str(p+1), current_player)))
        while player_input not in current_player:
            player_input = eval(input("\nInvalid Input \nEnter number again:"))
    else:
        player_input = current_player[0]
        print ("\nYour input is: %s" % player_input)
    return player_input
```

```python
def result_heli(player_input, player, p):
    #同上
    opponent = player[p==0]
    temp = []
    for i in range(len(opponent)):
        temp.append(player_input+opponent[i])
        temp.append(player_input*opponent[i])
        temp.append((player_input+opponent[i])%10)
        temp.append((player_input*opponent[i])%10)
    result = eval(input("\nOpponent's number: %s \nEnter your result:" % opponent))
    while result not in temp:
        result = eval(input("Invalid result \nEnter result again:"))
    result = result % 10 if result >=10 else result
    return result
```

```python
def player_round(player, p, count):
    
    player_input = input_heli(player, p, count)
    result = result_heli (player_input, player, p)
    
    if result == 0:
        player[p].remove(player_input)
    else: 
        player[p].append(result)
        player[p].remove(player_input)
    
    return player
```
